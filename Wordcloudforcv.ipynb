```python
arquivoPDF = pdfplumber.open('/content/Fernandacv.pdf')
```

```python
primeira_pagina = arquivoPDF.pages[0]

```

```python
textoCRU = primeira_pagina.extract_text()
textoCRU
```

```python
nltk.download('punkt')
```

```python
lista_de_palavras=nltk.tokenize.word_tokenize(textoCRU)
```

```python
lista_de_palavras
```

```python
lista_de_palavras = [palavra.lower()for palavra in lista_de_palavras]
lista_de_palavras
```

```python
pontuaçao = ['(',')',';','[',']',',']
```

```python
mes = ['(january)','(august)','(september)','(visa)','(https)','(sirva)','(documentation)']
```

```python
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('english')
stop_words
```

```python
keywords = [palavra for palavra in lista_de_palavras if not palavra in stop_words and not palavra in pontuaçao and not palavra in mes]
keywords
```

```python
stopwords = set(stop_words)
stopwords.update([''barbosa'',''mobility'',''brazil'',''july'',''january'',''phillip'',''morris'',''january'',''august'',''september'',''visa'',''https'',''sirva'',''documentation''])
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
textoCV = " ".join(s for s in keywords)
textoCV
```

```python
from wordcloud import WordCloud
```

```python
 wordcloud import WordCloud
import matplotpltlib.pyplot as 

word = WordCloud(
    background_color = '#0f54c9',
    max_font_size = 150,
    width = 1200,
    height=627,
    colormap = 'Blues'
).generate(textoCV)

#mostrar resultado
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(16,9))
ax.imshow(word)
ax.set_axis_off()
plt.imshow(word)
word.to_file('wordcloud.png')
plt.show()
```
